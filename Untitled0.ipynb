```python
import os
import sklearn
from sklearn.linear_model import LinearRegression
```

```python
# p6
data1=data.copy().dropna(subset=['Extra'])
m, b = np.polyfit(data1['Extra'],data1['Total'],1)
print(str(m)+"x +"+str(b))

linReg = LinearRegression()
linReg.fit(data1['Extra'].to_frame(),data1['Total'].to_frame())
print(str(linReg.coef_[0])+"x +"+str(linReg.intercept_))
```

```python
# p7
data1['predict']=linReg.predict(data1[['Extra']])
data1['error']=(data1['Total']-data1['predict'])**2
print(data1['error'].mean()**(.5))
```

```python
# p8
plt.scatter(data1['Extra'],data1['Total'])
xval = np.linspace(0, 5, 100).reshape(-1,1)
predictval=linReg.predict(xval)
plt.plot(xval,predictval,color='red')
```
